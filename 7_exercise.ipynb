```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

(train_images, _), (test_images, _) = mnist.load_data()

train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

train_images = train_images.reshape((len(train_images), 28, 28, 1))
test_images = test_images.reshape((len(test_images), 28, 28, 1))

def build_autoencoder(encoder_activation, decoder_activation):
    model = models.Sequential()

    model.add(layers.Conv2D(32, (3, 3), activation=encoder_activation, padding='same', input_shape=(28, 28, 1)))
    model.add(layers.MaxPooling2D((2, 2), padding='same'))
    model.add(layers.Conv2D(64, (3, 3), activation=encoder_activation, padding='same'))
    model.add(layers.MaxPooling2D((2, 2), padding='same'))

    model.add(layers.Conv2DTranspose(64, (3, 3), activation=decoder_activation, padding='same'))
    model.add(layers.UpSampling2D((2, 2)))
    model.add(layers.Conv2DTranspose(32, (3, 3), activation=decoder_activation, padding='same'))
    model.add(layers.UpSampling2D((2, 2)))
    model.add(layers.Conv2DTranspose(1, (3, 3), activation='sigmoid', padding='same'))

    return model

def train_model(encoder_activation, decoder_activation, epochs=10):
    model = build_autoencoder(encoder_activation, decoder_activation)
    model.compile(optimizer='adam', loss='binary_crossentropy')

    model.fit(train_images, train_images, epochs=epochs, batch_size=128, shuffle=True, validation_data=(test_images, test_images))

    return model

def generate_examples(model):
    reconstructed_images = model.predict(test_images)

    num_examples = 5
    plt.figure(figsize=(12, 4))

    for i in range(num_examples):
        plt.subplot(2, num_examples, i + 1)
        plt.imshow(test_images[i].reshape(28, 28), cmap='gray')
        plt.title('Original')

        plt.subplot(2, num_examples, i + num_examples + 1)
        plt.imshow(reconstructed_images[i].reshape(28, 28), cmap='gray')
        plt.title('Reconstructed')

    plt.show()

encoder_activations = ['relu', 'tanh', 'sigmoid']
decoder_activations = ['relu', 'tanh', 'sigmoid']

for encoder_activation in encoder_activations:
    for decoder_activation in decoder_activations:
        print(f"Training with Encoder Activation: {encoder_activation}, Decoder Activation: {decoder_activation}")
        model = train_model(encoder_activation, decoder_activation, epochs=5)
        generate_examples(model)
```
